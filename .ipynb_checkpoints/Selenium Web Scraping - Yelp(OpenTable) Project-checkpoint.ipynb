```python
chrome_driverpath = "C:/chromedriver-win64/chromedriver.exe"
service = Service(executable_path=chrome_driverpath)
driver2 = webdriver.Chrome(service=service) 
website2= "https://www.opentable.co.uk/metro/london-restaurants"
driver2.get(website2)
driver2.maximize_window()
time.sleep(4)
```
