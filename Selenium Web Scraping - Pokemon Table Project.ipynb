```python
# import relevant packages and modules
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import pygetwindow as gw
import time 
```

```python
chrome_driverpath = "C:/chromedriver-win64/chromedriver.exe"
service = Service(executable_path=chrome_driverpath)
driver1 = webdriver.Chrome(service=service) 
website1= "https://pokemondb.net/pokedex/all"
driver1.get(website1)
driver1.maximize_window()
time.sleep(1)
```

```python
#
try:
    button = driver1.find_element(By.CLASS_NAME, 'gdpr-accept')
    button.click()
except:
    pass
```

```python
# name
pokemon_name = driver1.find_elements(By.XPATH,'//tbody/tr/td[2]')

# type
pokemon_type = driver1.find_elements(By.XPATH,'//tbody/tr/td[3]')

# totalpoints
pokemon_tp = driver1.find_elements(By.XPATH,'//tbody/tr/td[4]')

# HP
pokemon_hp = driver1.find_elements(By.XPATH,'//tbody/tr/td[5]')

# attack
pokemon_attack = driver1.find_elements(By.XPATH,'//tbody/tr/td[6]')

# defense
pokemon_defence = driver1.find_elements(By.XPATH,'//tbody/tr/td[7]')

# sp_attack
pokemon_sp_attack = driver1.find_elements(By.XPATH,'//tbody/tr/td[8]')

# sp_defense
pokemon_sp_defense = driver1.find_elements(By.XPATH,'//tbody/tr/td[9]')

# speed
pokemon_speed = driver1.find_elements(By.XPATH,'//tbody/tr/td[10]')
```

```python
for i, name in zip(range(1,1215),pokemon_name[0:1215]):
    print(f"{i}. {name.text}")
```

#### Empty List

```python
pokemon_results = []
```

#### Dictionary

```python
for i in range(len(pokemon_name)):
    pokemon_dictionary = {'Name':pokemon_name[i].text,
                          'Type':pokemon_type[i].text,
                          'Total Points':pokemon_tp[i].text,
                          'Health Points': pokemon_hp[i].text,
                          'Attack': pokemon_attack[i].text,
                          'Defense': pokemon_defence[i].text,
                          'SP Attack': pokemon_sp_attack[i].text,
                          'SP Defense': pokemon_sp_defense[i].text,
                          'Speed': pokemon_speed[i].text
                         }
    pokemon_results.append(pokemon_dictionary)          
```

```python
pokemon_results
```

```python
pokemon_df = pd.DataFrame(pokemon_results)
pokemon_df
```

#### replacing unnecessary characters

```python
# replace \n with , or -
pokemon_df["Name"] = pokemon_df['Name'].apply(lambda x: x.replace('\n','-'))
pokemon_df["Type"] = pokemon_df['Type'].apply(lambda x: x.replace('\n',','))
```
